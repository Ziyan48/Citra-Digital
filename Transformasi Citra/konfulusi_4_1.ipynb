Mengambil library cv2, numpy dan matplotlib
cv2 = untuk membaca gambar atau vidio
numpy = untuk komputasi numerik
matplotlib = untuk menampilkan gambar di lingkungan notebook jupyter

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

img = membaca gambar "oye.jpg"
dstSkala = mengubah ukuran gambar img
plt.figure = membuat jendela dengan ukuran 10x5
plt.imshow = menampilkan gambar 
plt.title = menambahkan judul
plt.axis('off') = menyembunyikan sumbu 

```python
img = cv2.imread('oye.jpg')

dstSkala = cv2.resize(img, None, fx=2.5, fy=2, interpolation=cv2.INTER_CUBIC)
#cv2.imshow("img",img)
#cv2.imshow("dss",dstSkala)

plt.figure(figsize=(10, 5))
plt.imshow(img)
plt.title('Gambar Asli')
plt.axis('off')

plt.figure(figsize=(10, 5))
plt.imshow(dstSkala)
plt.title('Gambar Skala')
plt.axis('off')

```

print(img.shape) = membaca ukuran img.shape
baris, coloms, ghgh = img.shape = menyimpan jumlah baris, kolom ,dan chanel dari gambar

MTranslasi = np.float32([
     [2, 0, 100],

```python
# translasi

#import numpy as np

print(img.shape)

baris, coloms, ghgh = img.shape

MTranslasi = np.float32([
     [2, 0, 100],
     [0, 2, 50]           
    ])

print(MTranslasi, '\n')


dst = cv2.warpAffine(img, MTranslasi, (coloms, baris))

plt.figure(figsize=(10, 5))
plt.imshow(dst, cmap='gray')
plt.title('Gambar Translasi')
plt.axis('off')
```

```python
MRotasi = cv2.getRotationMatrix2D((coloms/2, baris/2), 90, 1)

print(MRotasi)

dstRotasi = cv2.warpAffine(img, MRotasi, (coloms, baris))

plt.figure(figsize=(10, 5))
plt.imshow(dstRotasi, cmap='gray')
plt.title('Gambar Rotasi')
plt.axis('off')
```

```python
# Affine

ttkAsal = np.float32([[172,109], [282,65], [272,316]])
ttkTujuan = np.float32([[53,128], [471,46], [275, 385]])

MAffine = cv2.getAffineTransform(ttkAsal, ttkTujuan)

print(MAffine)

dstAffine = cv2.warpAffine(img, MAffine, (coloms, baris))

plt.figure(figsize=(10, 5))
plt.imshow(dstAffine, cmap='gray')
plt.title('Gambar Rotasi')
plt.axis('off')
```
